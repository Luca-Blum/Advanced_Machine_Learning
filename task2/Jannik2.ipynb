```
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import umap
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```
x_train = pd.read_csv("data/X_train.csv",index_col=0,header = 0)
y_train = pd.read_csv("data/y_train.csv",index_col=0,header = 0)
x_test = pd.read_csv("data/X_test.csv",index_col=0,header = 0)
```

```
y_train['y']=y_train['y'].replace(2,0)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```
from scipy import stats
p=[]
for i in range(x_train.shape[1]):
  p.append(abs(stats.pearsonr(x_train[:,i], y_train['y'])[0]))
p.sort()
plt.plot(p)
plt.show()
```

```
p=[]
for i in range(x_train.shape[1]):
  p.append(abs(stats.spearmanr(x_train[:,i], y_train['y'])[0]))
p.sort()
plt.plot(p)
plt.show()
```

```
q=[value for value in p if value>0.45]
len(q)
```

```
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import balanced_accuracy_score
from sklearn.model_selection import GridSearchCV
```

```
kbest = SelectKBest(f_classif,k=106).fit(x_train,y_train.values.ravel())
X_train = kbest.transform(x_train)
X_test = kbest.transform(x_test)
```

```
reducer = umap.UMAP(random_state=36)
data = X_train
embedding = reducer.fit_transform(data)
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in y_train['y']])
plt.gca().set_aspect('equal', 'datalim')
plt.show()
```

```
#q=[value for value in p if value>0.08]
tester=[y_train['y'][i] for i in range(len(embedding[:,0])) if embedding[i,0]>7.0]
print(len(tester))
print(len([i for i in tester if i<1]))
```

```
Y_train=[y_train['y'][i] for i in range(len(embedding[:,0])) if embedding[i,0]<=7.0]
X_train=[x_train[i] for i in range(len(embedding[:,0])) if embedding[i,0]<=7.0]
```

```
p=[]
for i in range(len(X_train[0])):
  test=[]
  for j in range(len(X_train[:])):
    test.append(X_train[j][i])
  p.append(abs(stats.pearsonr(test, Y_train)[0]))
p.sort()
plt.plot(p)
plt.show()
```

```
p=[]
for i in range(len(X_train[0])):
  test=[]
  for j in range(len(X_train[:])):
    test.append(X_train[j][i])
  p.append(abs(stats.spearmanr(test, Y_train)[0]))
p.sort()
plt.plot(p)
plt.show()
```

```
kbest = SelectKBest(f_classif,k=106).fit(X_train,Y_train)
X_trainer = kbest.transform(X_train)
X_test = kbest.transform(x_test)

reducer = umap.UMAP(random_state=36)
data = X_trainer
embedding = reducer.fit_transform(data)
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in Y_train])
plt.gca().set_aspect('equal', 'datalim')
plt.show()
```
